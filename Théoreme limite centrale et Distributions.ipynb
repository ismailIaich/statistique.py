$\mathbf{POISSON}$: La probabilité d'observer $n$ succès à faible probabilité lorsque le nombre moyen de succès est de $\mu$
$$
P(n;\mu)=\frac{\mu^n e^{-\mu}}{n!}
$$

* esperance=$\mu$, variance=$\mu$
* Exemple : quelle est la probabilité d'observer 60 nouvelles mutations dans le génome d'un enfant, sachant que le nombre moyen de nouvelles mutations est de 50 ?

```python
u=3
n=range(0,15)
y=stats.poisson.pmf(n,u)
plt.plot(n,y,'o-')
plt.title('Poisson: $\mu$ =%i' % u,fontsize=15)
plt.xlabel('n',fontsize=15)
plt.ylabel('Probabilité de n',fontsize=15)
plt.show()
```

$\mathbf{EXPONENTIELLE}$:
$P(x;k)=k e^{-kx}, \hspace{1in} x \in [0,\infty], \hspace{0.1in} k>0$

* esperance=$1/k$,  variance=$1/k^2$
* Cette distribution décrit les intervalles de temps dans un processus homogène de Poisson.

```python
k=0.4
x=np.arange(0,15,0.1)
y=k*np.exp(-k*x)
plt.plot(x,y,'-')
plt.title('Exponential: $k$ =%.2f' % k,fontsize=15)
plt.xlabel('x',fontsize=15)
plt.ylabel('Probabilité de densité',fontsize=15)
plt.show()
```

$\mathbf{BETA}$: $P(x;\alpha,\beta)=\displaystyle \frac{\Gamma(\alpha+\beta)}{\Gamma(\alpha) \Gamma(\beta)}x^{\alpha-1}(1-x)^{\beta-1}$, $\hspace{1in} x \in  [0;1], \hspace{0.1in} \alpha>0, \hspace{0.1in}\beta>0$

* esperance=$\displaystyle \frac{\alpha}{\alpha+\beta}$, variance=$\displaystyle \frac{\alpha \beta}{(\alpha + \beta)^2(\alpha + \beta +1)}$
* Cette distribution apparaît souvent en génétique des populations et dans l'inférence bayésienne.

```python
a=0.5
b=0.5
x=np.arange(0.01,1,0.01)
y=stats.beta.pdf(x,a,b)
plt.plot(x,y,'-')
plt.title('Beta: a=%.1f, b=%.1f' % (a,b) , fontsize=15)
plt.xlabel('x',fontsize=15)
plt.ylabel('Probabilité de densité',fontsize=15)
plt.show()
```

## Qu'est-ce que le théorème de la limite centrale (TLC) ?

* Le théorème de la limite centrale (TLC) est une théorie statistique qui stipule qu'étant donné une taille d'échantillon suffisamment grande à partir d'une population avec un niveau fini de variance, la moyenne de tous les échantillons de la même population sera approximativement égale à la moyenne de la population. 

En outre, tous les échantillons suivront un modèle de distribution approximativement normal, toutes les variances étant approximativement égales à la variance de la population divisée par la taille de chaque échantillon.

Selon le théorème de la limite centrale, la moyenne d'un échantillon de données se rapprochera de la moyenne de l'ensemble de la population en question à mesure que la taille de l'échantillon augmentera, indépendamment de la distribution réelle des données, qu'elle soit normale ou non. 

En règle générale, les tailles d'échantillon égales ou supérieures à 30 sont considérées comme suffisantes pour que le théorème de la limite centrale s'applique, ce qui signifie que la distribution des moyennes de l'échantillon est relativement normale.


* Ce qui est étonnant, c'est que ce théorème affirme qu'une distribution normale apparaît quelle que soit la distribution initiale. 
* Même si notre population présente une distribution asymétrique, ce qui se produit lorsque nous examinons des éléments tels que les revenus ou le poids des personnes, une distribution d'échantillonnage pour un échantillon de taille suffisamment grande sera normale.
* Une distribution de population qui est asymétrique (même très fortement asymétrique) a des applications très importantes dans la pratique statistique.

## Pourquoi TLC est important :

* De nombreuses pratiques statistiques, telles que celles impliquant des tests d'hypothèse ou des intervalles de confiance, reposent sur des hypothèses concernant la population à partir de laquelle les données ont été obtenues.
* L'une des hypothèses formulées initialement dans un cours de statistiques est que les populations avec lesquelles nous travaillons sont normalement distribuées.
* L'hypothèse selon laquelle les données proviennent d'une distribution normale simplifie les choses mais semble un peu irréaliste.
* Il suffit de travailler un peu avec des données réelles pour constater que les valeurs aberrantes, l'asymétrie, les pics multiples et l'asymétrie apparaissent assez régulièrement. 
* Il est possible de contourner le problème des données provenant d'une population qui n'est pas normale. 
* L'utilisation d'une taille d'échantillon appropriée et le théorème de la limite centrale nous aident à contourner le problème des données provenant de populations qui ne sont pas normales.

### Attention :

Pour que les conclusions du théorème soient valables, il faut que la taille de l'échantillon soit suffisamment grande. 

* En règle générale, un échantillon de 30 personnes est considéré comme suffisamment grand.

L'analyse exploratoire des données peut nous aider à déterminer la taille de l'échantillon nécessaire pour une situation donnée.

Examinons maintenant l'interface de programmation à travers quelques exemples.

Nous commençons par quelques bibliothèques standard.

```python
from functools import partial # permet de définir une fonction avec des arguments partiels

N=1000 # nombre de fois que n échantillons sont prélevés. Essayez de faire varier ce nombre.
nobb=101 # nombre de limites d'emplacements sur les placettes
n=np.array([1,2,3,5,10,100, 200]) # nombre d'échantillons sur lesquels la moyenne est calculée

exp_mean=3 # esperance de la distribution exponentielle 
a,b=0.7,0.5 # parametre de la distribution beta 

dist=[partial(np.random.random),partial(np.random.exponential,exp_mean),partial(np.random.beta,a,b)]
title_names=["Flat", "Exponential (mean=%.1f)" % exp_mean, "Beta (a=%.1f, b=%.1f)" % (a,b)]
drange=np.array([[0,1],[0,10],[0,1]]) # rang de la distribution
means=np.array([0.5,exp_mean,a/(a+b)]) # esperances des distributions
var=np.array([1/12,exp_mean**2,a*b/((a+b+1)*(a+b)**2)]) # variances des distributions

binrange=np.array([np.linspace(p,q,nobb) for p,q in drange])
ln,ld=len(n),len(dist)
plt.figure(figsize=((ld*4)+1,(ln*2)+1))

for i in range(ln): #  boucle sur le nombre d'échantillons n pour calculer la moyenne
    for j in range(ld): # boucle sur les differents distributions
        plt.subplot(ln,ld,i*ld+1+j)
        plt.hist(np.mean(dist[j]((N,n[i])),1),binrange[j],normed=True)
        plt.xlim(drange[j])
        if j==0:
            plt.ylabel('n=%i' % n[i],fontsize=15)        
        if i==0:
            plt.title(title_names[j], fontsize=15)
        else:
            clt=(1/(np.sqrt(2*np.pi*var[j]/n[i])))*np.exp(-(((binrange[j]-means[j])**2)*n[i]/(2*var[j])))
            plt.plot(binrange[j],clt,'y',linewidth=2)     
plt.show()
```

Dans les graphiques ci-dessus, la courbe jaune est la distribution gaussienne prédite à partir de Thereome de la limite centrale. Remarquez que le taux de convergence de la moyenne de l'échantillon vers la gaussienne dépend de la distribution mère originale. En outre, la moyenne de la distribution gaussienne dépend de la distribution mère originale, 

- l'esperance de la distribution gaussienne est la même que celle de la distribution mère originale,
- la largeur de la distribution gaussienne est égale à $1/\sqrt{n}$.
